# Data Visualizations

```python
fig = px.treemap(df, path=["Platform", "Genre"], title="Grafik Treemap Hierarki Platform dan Genre")
fig.show()
```

Dari sini bisa kita dapat melihat distribusi genre video games di setiap platformnya. Dapat dilihat bahwa game dengan genre action dan sports mendominasi di banyak platform.



```python
# exclude top 10 platform
value_counts = df['Platform'].value_counts()
to_remove = value_counts[value_counts >= 600].index
df2 = df[~df.Platform.isin(to_remove)]

fig = px.treemap(df2, path=["Platform", "Genre"], title="Grafik Treemap Hierarki Platform dan Genre (di luar platform top 10)")
fig.show()
```

Dari grafik treemap di atas, kita dapat melihat lebih jelas distribusi genre video games berdasarkan platformnya untuk platform di bawah top 10 dengan judul game yang dirilis terbanyak.


```python
df1 = df
df1['count'] = df.groupby("Genre")["Platform"].transform(len)
fig = px.parallel_categories(df1, dimensions=["Platform", "Genre"], color = "count",
                             color_continuous_scale=px.colors.sequential.Inferno,
                             title="Diagram Kategori Paralel")
fig.show()
```

Grafik di atas menunjukkan hubungan sekaligus perbandingan data. Garis yang ditampilkan menunjukkan hubungan data sedangkan warna dan ukuran kotak menunjukkan kuantitas dari data yang terhubung. Dapat diketahui bahwa game dengan genre action memiliki judul game yang dirilis terbanyak dan hubungannya dengan platform yang merilis game tersebut

```python
# visualize the distribution of games released
fig = px.histogram(df, x = "Year", title = "Grafik Persebaran Jumlah Game yang Dirilis Setiap Tahun")
fig.show()
```

Dari grafik persebaran di atas, dapat kita ketahui bahwa jumlah game yang dirilis tertinggi terjadi pada tahun 2008 sampai 2010 dan jumlah game yang dirilis terendah terjadi pada tahun 1983 sampai 1990

```python
fig = px.line(df.groupby(["Year"])["Global_Sales"].sum().reset_index(name="Sales_Sum"), x = "Year", y =["Sales_Sum"],
              title="Grafik Penjualan Game Global Setiap Tahun", labels={"Year":"Tahun", "value":"Jumlah Penjualan Game GLobal (dalam juta)",
              "Sales_Sum":"Jumlah Penjualan Game Global"})
fig.show()
```

```python
df_sum = df.groupby(["Year"]).sum().reset_index()
fig = px.area(title="Grafik Penjualan Game Global Setiap Tahun",
              labels={"Year":"Tahun", "value":"Jumlah Penjualan Game GLobal (dalam juta)", "Sales_Sum":"Jumlah Penjualan Game Global"})

fig.add_trace(go.Scatter(
    name = 'NA Sales',
    x = df_sum["Year"],
    y = df_sum["NA_Sales"],
    stackgroup='one'
   ))
fig.add_trace(go.Scatter(
    name = 'EU Sales',
    x = df_sum["Year"],
    y = df_sum["EU_Sales"],
    stackgroup='one'
   ))
fig.add_trace(go.Scatter(
    name = 'JP Sales',
    x = df_sum["Year"],
    y = df_sum["JP_Sales"],
    stackgroup='one'
   ))
fig.add_trace(go.Scatter(
    name = 'Other Sales',
    x = df_sum["Year"],
    y = df_sum["Other_Sales"],
    stackgroup='one'
   ))
fig.show()
```

Dari grafik area di atas, dapat kita lihat bahwa distribusi penjualan game setiap tahunnya di masing-masing area mencapai tingkat tertingginya di tahun 2005 sampai dengan tahun 2010.

```python
df1 = df.groupby("Platform")["Global_Sales"].sum().reset_index(name="Sales_Sum").sort_values(by="Sales_Sum", ascending = True)
fig = px.histogram(df1, x = "Sales_Sum", y = "Platform", orientation="h", title="Jumlah Penjualan Game Global Berdasarkan Platform", labels={"Sales_Sum":"Penjualan (dalam jutaan)"})
fig.show()
```

```python
# visualize ratio of games based on the platform
fig = px.pie(df.groupby("Platform")["Platform"].count().reset_index(name="count"), values = "count", names = "Platform", title = "Pie Chart Jumlah Game yang Dirilis Pada Setiap Platform")
fig.show()
```

Dari pie chart di atas, dapat diketahui dengan jelas perbandingan jumlah game yang dirilis pada setiap platform dengan persentase yang ditampilkan. Terlihat bahwa platform dengan jumlah judul game yang dirilis terbanyak adalah DS disusul oleh PS2.

```python
# visualize ratio and total games released from every publishers
df1 = df.groupby("Publisher")["Publisher"].count().reset_index(name="count")
df1 = df1.loc[df1["count"] > .02*df["Publisher"].count()]
df2 = df.groupby("Publisher")["Publisher"].count().reset_index(name="count")
df2 = df2.loc[df2["count"] <= .02*df["Publisher"].count()]
df2 = pd.DataFrame({"Publisher":["Other"], "count":[df2["count"].sum().sum()]})
TGByPublisher = pd.concat([df1, df2], axis=0)

fig = px.pie(TGByPublisher, values="count", names="Publisher", title="Perbandingan Jumlah Game yang Dirilis Berdasarkan Publisher")
fig.show()
```

Dari pie chart di atas, dapat diketahui dengan jelas perbandingan jumlah game yang dirilis oleh setiap publisher dengan persentase yang ditampilkan. Terlihat bahwa publisher dengan jumlah judul game yang dirilis terbanyak adalah Electronic Arts (EA). Dengan mempertimbangkan persentase yang tertampil, dapat diketahui bahwa jumlah judul game yang dirilis oleh publisher lebih menyebar merata.

```python
fig = px.bar(df.groupby("Genre")["Genre"].count().reset_index(name="count"), x = "Genre", y = "count",
             title="Grafik Jumlah Game yang Rilis Berdasarkan Genre", labels={"count":"Jumlah Game"})
fig.show()
```

Dari grafik bar chart dia atas dapat kita ketahui bahwa game dengan genre action adalah game dengan jumlah game yang dirilis terbanyak sedangkan game dengan genre puzzle adalah game dengan jumlah game yang dirilis paling sedikit.

```python
fig = px.bar(df.groupby("Genre")["Global_Sales"].sum().reset_index(name="Sales"), x = "Genre", y = "Sales",
             title="Grafik Penjualan Game Global Berdasarkan Genre", labels={"Sales":"Penjualan (dalam jutaan)"})
fig.show()
```

Pada grafik bar chart penjualan, dapat kita ketahui bahwa game dengan genre action adalah game dengan jumlah penjualan global terbanyak sedangkan game dengan genre strategy adalah game dengan jumlah penjualan global paling sedikit. Bisa dilihat bahwa terdapat beberapa perbedaan dibandingkan dengan grafik bar chart jumlah game yang dirilis.

```python
df_sum = df.groupby(["Genre"]).sum().reset_index()
fig = go.Figure(
    data=[
        go.Bar(
          name = 'NA Sales',
          x = df_sum["Genre"],
          y = df_sum["NA_Sales"],
          offsetgroup='one'
        ),
        go.Bar(
          name = 'EU Sales',
          x = df_sum["Genre"],
          y = df_sum["EU_Sales"],
          offsetgroup='one',
          base = df_sum["NA_Sales"]
        ),
        go.Bar(
          name = 'JP Sales',
          x = df_sum["Genre"],
          y = df_sum["JP_Sales"],
          offsetgroup='one',
          base = df_sum["NA_Sales"] + df_sum["EU_Sales"]
        ),
        go.Bar(
          name = 'Other Sales',
          x = df_sum["Genre"],
          y = df_sum["Other_Sales"],
          offsetgroup='one',
          base = df_sum["NA_Sales"] + df_sum["EU_Sales"] + df_sum["JP_Sales"]
        )
    ],
    layout=go.Layout(
        title="Grafik Penjualan Game Global Berdasarkan Genre",
        yaxis_title="Penjualan (dalam jutaan)"
    )
)

fig.show()
```

Dari grafik bar chart dia atas dapat kita ketahui bahwa game dengan genre action adalah game dengan jumlah penjualan global terbanyak sedangkan game dengan genre strategy adalah game dengan jumlah penjualan global paling sedikit. Grafik diatas adalah grafik interaktif sehingga persebaran data dapat diketahui dengan meng-hover bagian box pada grafik.

```python
Count_Corr = df.groupby(["Publisher"])["Global_Sales"].sum().reset_index(name="Sales")
Sales_Corr = df.groupby(["Publisher"])["Global_Sales"].count().reset_index(name="Count")
Temp = pd.concat([Count_Corr, Sales_Corr["Count"]], axis=1)
fig = px.scatter(Temp, x = "Count", y="Sales", color="Publisher", trendline="ols", trendline_scope="overall",
                 labels={"Sales":"Jumlah Penjualan Game Yang Dirilis (dalam juta)", "Count":"Jumlah Judul Game Yang Dirilis"})
fig.update_layout(title="Grafik Scatter Jumlah Dan Penjualan Game Yang Dirilis Berdasarkan Publisher")
fig.show()
```

Dari grafik scatter jumlah dan penjualan game yang dirilis berdasarkan publisher di atas, dapat diketahui bahwa jumlah judul game cenderung berbanding lurus dengan jumlah penjuala game yang dirilis dengan nilai koefisien determinasi (R^2) bernilai 0,639834.

```python
Count_Corr = df.groupby(["Genre"])["Global_Sales"].sum().reset_index(name="Sales")
Sales_Corr = df.groupby(["Genre"])["Global_Sales"].count().reset_index(name="Count")
Temp = pd.concat([Count_Corr, Sales_Corr["Count"]], axis=1)
fig = px.scatter(Temp, x = "Count", y="Sales", color="Genre", trendline="ols", trendline_scope="overall",
                 labels={"Sales":"Jumlah Penjualan Game Yang Dirilis (dalam juta)", "Count":"Jumlah Judul Game Yang Dirilis"})
fig.update_layout(title="Grafik Scatter Jumlah Dan Penjualan Game Yang Dirilis Berdasarkan Genre")
fig.show()
```

Dari grafik scatter jumlah dan penjualan game yang dirilis berdasarkan genre di atas, dapat diketahui lebih jelas bahwa jumlah judul game cenderung berbanding lurus dengan jumlah penjuala game yang dirilis dengan nilai koefisien determinasi (R^2) bernilai 0,790709.
